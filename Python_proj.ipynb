```python
#derive new columns discount, sale price and profit
df['discount']=df['list_price']*df['discount_percent']*0.01
df['sale_price']=df['list_price']-df['discount']
df['profit']=df['sale_price']-df['cost_price']
df['profit_percentage']=((df['profit']/df['cost_price'])*100).round(2)
df.head(5)
```

```python
import pandas as pd
import sqlalchemy as sal
import pymysql
from sqlalchemy.types import Integer, Float, String, Date

dtype_mapping = {
    'int64': Integer(),
    'float64': Float(),
    'object': String(30),
    'datetime64[ns]': Date()
}

# Function to map pandas dtypes to SQLAlchemy dtypes
def map_dtypes(df, dtype_mapping):
    sql_dtypes = {}
    for column, dtype in df.dtypes.items():
        sql_dtypes[column] = dtype_mapping.get(str(dtype), String(255))
    return sql_dtypes

# Replace with your MySQL credentials
db_user = 'root'
db_password = 'Password'
db_host = 'localhost'  # e.g., 'localhost' or an IP address
db_name = 'python_retailer_jyp_nb'

# Assuming df is your DataFrame
df['order_date'] = pd.to_datetime(df['order_date'], format='%Y-%m-%d')

# Create the MySQL engine
engine = sal.create_engine(f'mysql+pymysql://{db_user}:{db_password}@{db_host}/{db_name}')

# Optional: Test the connection
try:
    connection = engine.connect()
    print("Connection to MySQL successful!")
    connection.close()  # Close the connection after testing
except Exception as e:
    print(f"Connection failed: {e}")

table_name = 'retailer_data'  # Replace with your table name

# Map data types and load DataFrame into the MySQL table
sql_dtypes = map_dtypes(df, dtype_mapping)
df.to_sql(table_name, con=engine, if_exists='replace', index=False, dtype=sql_dtypes)  # Corrected 'dtype' argument

print(f"DataFrame successfully loaded into the '{table_name}' table with replace option.")
```
